```
# Null Hypothesis: is there any relationship between two categorical values

Alternate Hypothesis: There is a relationship between two categorical values

Questions we are going to answer through statistical test:

is there any relationship number of births with respect to gender?
is there any relationship number of births with respect to year wise?

chi square test: shows significance of relationship, significance to the target.
fischers exact test: used for 2x2 matrix

statistics alone cannot prove anything.All the tests we do are based on relative likelihood.
contingency table: also called as crosstab.used to summerise the several categorical values.
the more the chisquare value is the more the probability that the var is statistically significance var.
bydefaultly chisqaure table gives the p,dof,chi2.

chisquare stats=sum(obsrvd-expectd)^2/expected
```

```
df["births"].value_counts()
```

```
df["gender"].value_counts()
```

```
from scipy import stats
P = pd.crosstab(df["gender"],df["births"])
print(P)
```

```
from scipy.stats import chi2_contingency
obs = P
chi2_contingency(obs,correction=True)
```



P-value is less than .05 so our alternative hypothesis is true that is birth and gender are related
